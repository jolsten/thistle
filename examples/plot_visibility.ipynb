# Visibility Circles

Plot ground-site visibility rings at 0°, 5°, and 10° minimum elevation for a 500 km satellite.

```python
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import matplotlib.pyplot as plt

from thistle.ground_sites import visibility_circle

SAT_ALT = 500_000
ELEVATIONS = [(0, "C0"), (5, "C1"), (10, "C2")]
```

## Continental US ground site

```python
LAT, LON = 40.0, -105.0

fig, ax = plt.subplots(subplot_kw={"projection": ccrs.PlateCarree()})
ax.set_global()
ax.coastlines(linewidth=0.5)
ax.add_feature(cfeature.BORDERS, linewidth=0.3)
ax.gridlines(draw_labels=True, linewidth=0.3, alpha=0.5)

for el, color in ELEVATIONS:
    lats, lons = visibility_circle(LAT, LON, 0.0, sat_alt=SAT_ALT, min_el=el)
    ax.plot([*lons, lons[0]], [*lats, lats[0]], color=color, label=f"{el}°",
            transform=ccrs.Geodetic())

ax.plot(LON, LAT, "k+", markersize=10, transform=ccrs.PlateCarree())
ax.legend(title="Min elevation", loc="lower left")
ax.set_title(f"Boulder, CO \u2013 {SAT_ALT/1000:.0f} km altitude")
plt.tight_layout()
plt.show()
```

## Antimeridian crossing

A ground site near 180° longitude produces a visibility circle that wraps across the antimeridian.

```python
LAT, LON = -39.26, 177.87

fig, ax = plt.subplots(subplot_kw={"projection": ccrs.PlateCarree()})
ax.set_global()
ax.coastlines(linewidth=0.5)
ax.add_feature(cfeature.BORDERS, linewidth=0.3)
ax.gridlines(draw_labels=True, linewidth=0.3, alpha=0.5)

for el, color in ELEVATIONS:
    lats, lons = visibility_circle(LAT, LON, 0.0, sat_alt=SAT_ALT, min_el=el)
    ax.plot([*lons, lons[0]], [*lats, lats[0]], color=color, label=f"{el}°",
            transform=ccrs.Geodetic())

ax.plot(LON, LAT, "k+", markersize=10, transform=ccrs.PlateCarree())
ax.legend(title="Min elevation", loc="lower left")
ax.set_title(f"Mahia Peninsula (°E) \u2013 {SAT_ALT/1000:.0f} km altitude")
plt.tight_layout()
plt.show()
```
